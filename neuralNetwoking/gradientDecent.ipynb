```python
import numpy as np
```

```python
points = []
true_w =np.array([1,2,3,4,5])
D = len(true_w)

for i in range (50000):
    x = np.random.randn(D)
    y = true_w.dot(x) + np.random.randn()
    points.append((x,y))
# points
```

```python
points
```

```python
def Loss(w):
    return sum((w.dot(x)-y)**2 for x ,y in points)/len(points)

def dLoss(w):
    return sum(2*(w.dot(x)-y)*x for x,y in points ) /len(points)

def GradientDescent(Loss,dLoss,D):
    w = np.zeros(D)
    lr = 0.01
    
    for i in range(1000):
        loss = Loss(w)
        Gradient = dLoss(w)
        w = w -lr*Gradient
        print(f'Iterations {i}: w = {w},L(w) = {loss}')

GradientDescent(Loss,dLoss,D)
```

```python
def sloss(w,i):
    x,y = points[i]
    return (w.dot(x)-y)**2

def sDloss(w,i):
    x,y = points[i]
    return (2*(w.dot(x)-y)*x)

def sGradientDescent(sloss,sDloss,D,numPoints):
    w = np.zeros(D)
    numUpdates = 1
    for t in range(1000):
        for i in range (numPoints):
            
            loss = sloss(w,i)
            gradient = sDloss(w,i)
            numUpdates +=1
            lr = 1/numUpdates
            w = w - lr*gradient
        print(f'Iterations {t}: w = {w},L(w) = {loss}')

sGradientDescent(sloss,sDloss,D,len(points))


    
```
