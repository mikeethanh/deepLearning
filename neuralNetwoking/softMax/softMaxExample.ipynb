```python
learningRate = 0.01
losses=[]
num_iter = 2000
for epoch in range (num_iter):
    z = x.dot(theta)
    exp_z = np.exp(z)
    y_hat = exp_z / np.sum(exp_z,axis=1,keepdims=True)
    loss = -np.log(np.sum(y_hat*y,axis=1))
    losses.append(np.sum(loss))
    dz = y_hat - y
    dtheta = X.T.dot(dz)
    theta = theta - learningRate*dtheta 
```
